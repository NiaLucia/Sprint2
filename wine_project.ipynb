```python
red_wine = pd.read_csv('Wine Project\winequality-red.csv', delimiter= ';')

white_wine = pd.read_csv('Wine Project\winequality-white.csv', delimiter= ';')
```

```python
red_wines = pd.DataFrame(red_wine)
white_wines = pd.DataFrame(white_wine)

red_wines = round(red_wines, 2)
white_wines = round(white_wines, 2)
```

```python
# Let's now group our qualities into 3 categories: low, medium and high quality wines.
# we are creating a new column called "quality_label", we define a range and associate that range with a label


red_wines['quality_label'] = red_wines['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')
```

```python
# here we are transforming these labels into categrical data type (specific to pandas) instead of simple string

red_wines['quality_label'] = pd.Categorical(red_wines['quality_label'],
categories=['low', 'medium', 'high'])
```

```python
white_wines['quality_label'] = white_wines['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

white_wines['quality_label'] = pd.Categorical(white_wines['quality_label'],
categories=['low', 'medium', 'high'])
```

```python
# Epic 3: Exploratory Analysis

pd.concat([red_wines.describe().round(2), white_wines.describe().round(2)], axis=1, keys=["Red Wines Stats", "White Wines Stats"])

```

```python
# comments:
# the mean residual sugar in red wine is 2.54 and in white 6.39 --> more sugar means more ripened or warm climate
# the mean of total sulfur dioxide in red is 46.47 and in white 138.36
# red wine has more sulphates (0.66) than white (0.49)
# the volatile acidity of red is higher (0.53) than white (0.28)
```
